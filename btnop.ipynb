```python
# bai 2
n=int(input())
lst=[]
for i in range(2,n+1):
    while n%i==0:
        lst.append(i)
        n=n//i
print(f'phan tich n ra thua so nguyen to la:{lst}')

    
```

```python
# bai 1
a,b,c=map(int,input().split())
if a==b==c:
    print('tam giac deu')
elif a**2 +b**2==c**2 or a**2 +c**2==b**2 or c**2 +b**2==a**2:
    print('tam giac vuong')
elif (a==b and a+b>c ) or (b==c and b+c>a) or (a==c and a+c>b):
    print('tam giac can')
elif a+b>c and b+c>a and a+c>b:
    print('tam giac thuong')
```

```python
#bai 5
n=int(input())
giaithua=1
for i in range(1,n+1):
    giaithua*=i
dem=0
while giaithua%10==0:
    dem+1
    giaithua//=10
print(f'so chu so khong tan cung la: {giaithua%10}')
```
