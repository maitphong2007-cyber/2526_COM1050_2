```python
# tow sum
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        for i in range(1,len(nums)):
            for k in range(0,i):
                s=[]
                if nums[i]+nums[k]==target:
                    s.append(k)
                    s.append(i)
                    return s
```

```python
#. Longest Substring Without Repeating Characters
class Solution:
    def lengthOfLongestSubstring(self, s: str) -> int:
        s=list(map(str,s))
        dai=[]
        if s==[]:
            return 0
        for i in range(0,len(s)):
                lst=[]
                for k in range(i,len(s)):
                    if k==len(s)-1 and s[k] not in lst:
                        dai.append(k-i+1)
                    if s[k] not in lst:
                        lst.append(s[k])
                    else:
                        dai.append(k-i)
                        break
        return max(dai)
```

```python
#Zigzag Conversion
class Solution:
    def convert(self, s: str, numRows: int) -> str:
        xau=''
        if numRows==1:
            return s
        else:
            for i in range(0,numRows):
                if i==0 or i==numRows-1:
                    n=i
                    while n<len(s):
                        xau+=s[n]
                        n+=2*numRows-2
                else:
                    n=i
                    z=(numRows - i -1)*2
                    while n<len(s):
                        xau+=s[n]
                        if n+z<len(s):
                            xau+=s[n+z]
                        n+=2*numRows-2
            return xau
```

```python
#Reverse Integer
class Solution:
    def reverse(self, x: int) -> int:
        if 0<x<=2**31-1:
            x=str(x)
            if int(x[::-1])<=2**31-1:
                return int(x[::-1])
            else:
                return 0
        elif x==0:
            return 0
        elif -2**31<=x:
            x=-1*x
            x=str(x)
            if int(x[::-1])*-1>=-2**31:
                return int(x[::-1])*-1
            else:
                return 0
```

```python
#Palindrome Number
class Solution:
    def isPalindrome(self, x: int) -> bool:
        if x<0:      
            return False 
        else:
            x=str(x)
            if int(x)==int(x[::-1]):
                return True
            else:
                return False
```

```python
#Container With Most Water
class Solution:
    def maxArea(self, height: List[int]) -> int:
        left=0
        right=len(height)-1
        dientich=min(height[left],height[right])*(right-left)
        while left<right:
            dientich=max(min(height[left],height[right])*(right-left),dientich)
            if height[left]<height[right]:
                left+=1
            elif height[left]>height[right]:
                right-=1
            else:
                left+=1
                right-=1
        return dientich
```

```python
 #Integer to Roman
class Solution:
    def intToRoman(self, num: int) -> str:
        xau=''
        while num//1000!=0:
            xau+=num//1000*'M'
            num%=1000
        while num//100!=0:
            so=num//100
            if so<=3:
                xau+=so*'C'
            elif 3<so<=5:
                xau+=(5-so)*'C'+'D'
            elif 5<so<=8:
                xau+='D'+(so-5)*'C'
            else:
                xau+=(10-so)*'C'+'M'
            num%=100
        while num//10!=0:
            so=num//10
            if so<=3:
                xau+=so*'X'
            elif 3<so<=5:
                xau+=(5-so)*'X'+'L'
            elif 5<so<=8:
                xau+='L'+(so-5)*'X'
            else:
                xau+=(10-so)*'X'+'C'
            num%=10
        while num>=1:
            so=num
            if so<=3:
                xau+=so*'I'
            elif 3<so<=5:
                xau+=(5-so)*'I'+'V'
            elif 5<so<=8:
                xau+='V'+(so-5)*'I'
            else:
                xau+=(10-so)*'I'+'X'
            num-=so
        
        return xau
```

```python
#Roman to Integer
class Solution:
    def romanToInt(self, s: str) -> int:
        xau=s
        dic={'I':1,'V':5,'X':10,'L':50,'C':100,"D":500,"M":1000}
        tong=0
        tong+=dic[xau[len(xau)-1]]
        for i in range(1,len(xau)):
            if dic[xau[i]]>dic[xau[i-1]]:
                tong-=dic[xau[i-1]]
            elif dic[xau[i]]<dic[xau[i-1]]:
                tong+=dic[xau[i-1]]
            elif dic[xau[i]]==dic[xau[i-1]]:
                tong+=dic[xau[i-1]]
        return tong
```

```python
#3 sum
class Solution:
    def threeSum(self, nums: List[int]) -> List[List[int]]:
        nums.sort()
        kq=[]
        for i in range(0,len(nums)-2):
            if nums[i]>0:
                break
            elif i>0 and nums[i]==nums[i-1]:
                continue
            else:
                left=i+1
                right=len(nums)-1
                while left<right:
                        if nums[i]+nums[right]+nums[left]<0:
                            left+=1
                        elif nums[i]+nums[right]+nums[left]>0:
                            right-=1
                        elif nums[i]+nums[right]+nums[left]==0:
                            kq.append([nums[i],nums[left],nums[right]])
                            while left<right and nums[left]==nums[left+1]:
                                left+=1
                            while left<right and nums[right]==nums[right-1]:
                                right-=1
                            left+=1
                            right-=1
        return kq

```

```python
# 3 sum closet
class Solution:
    def threeSumClosest(self, nums: List[int], target: int) -> int:
        nums.sort()
        tong=float('inf')
        for i in range(0,len(nums)-2):
            left=i+1
            right=len(nums)-1
            while left< right:
                if nums[i]+nums[left]+nums[right]==target:
                    return target
                elif nums[i]+nums[left]+nums[right]<target:
                    if abs(nums[i]+nums[left]+nums[right]-target)<abs(tong-target):
                        tong=nums[i]+nums[left]+nums[right]
                    left+=1
                elif nums[i]+nums[left]+nums[right]>target:
                    if abs(nums[i]+nums[left]+nums[right]-target)<abs(tong-target):
                        tong=nums[i]+nums[left]+nums[right]
                    right-=1
        return tong
```

```python
# 4 sum
class Solution:
    def fourSum(self, nums: List[int], target: int) -> List[List[int]]:
        nums.sort()
        kq=[]
        for i in range(0,len(nums)-3):
            if i!=0 and nums[i]==nums[i-1]:
                continue
            else:
                for k in range(i+1,len(nums)-2):
                    tong=target-nums[i]-nums[k]
                    if nums[k]==nums[k-1] and k!=i+1:
                        continue
                    else:
                        left= k+1
                        right=len(nums)-1
                        while left<right:
                            if (nums[left]==nums[left-1] and left!=k+1) :
                                left+=1
                            elif nums[left]+nums[right]==tong:
                                lst=[nums[i],nums[k],nums[left],nums[right]]
                                kq.append(lst)
                                left+=1
                                right-=1
                            elif nums[left]+nums[right]<tong:
                                left+=1
                            else:
                                right-=1
        return kq            
```
